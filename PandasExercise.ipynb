# Ex1 - Filtering and Sorting Data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_table(url,sep = '\t')
chipo.head(10)
```

```python
chipo.info()
```

### Step 4. How many products cost more than $10.00?

```python
# Check if chipo['item_price'] is a string
print(chipo['item_price'].dtype)

# Remove dollar sign and convert to float
chipo['item_price'] = chipo['item_price'].str.replace('[\$,]', '', regex=True).astype(float)

# Now filter items costing more than $10
more_than_ten = chipo.loc[chipo['item_price'] > 10]
print(more_than_ten)
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
item_prices = chipo[['item_name', 'item_price']]
item_prices
```

### Step 6. Sort by the name of the item

```python
sorted_item_prices = item_prices.sort_values(by='item_name')
print(sorted_item_prices[['item_name', 'item_price']])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
most_expensive_item = chipo.loc[chipo['item_price'].astype(float).idxmax()]

item_name = most_expensive_item['item_name']
quantity_most_expensive = most_expensive_item['quantity']

print("Quantity of the most expensive item ordered:", item_name + ' at $' + str(quantity_most_expensive))
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python

veggie_salad_bowl_count = chipo[chipo['item_name'] == 'Veggie Salad Bowl']['quantity'].sum()

print("Veggie Salad Bowl was ordered:", veggie_salad_bowl_count, "times.")
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
canned_soda_orders = chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)]

canned_soda_count = canned_soda_orders.shape[0]

print("Canned Soda was ordered more than once:", canned_soda_count, "times.")
```
