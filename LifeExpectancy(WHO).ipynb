# **Data Cleaning**

```python
df1['population'].value_counts()
```

```python
# Define a list of column names where missing values will be filled with the mean of the column
numerical_data_mean = [
    'life_expectancy_',
    'total_expenditure',
    'income_composition_of_resources',
    'schooling'
]

# Define a list of column names where missing values will be filled with the median of the column
numerical_data_median = [
    '_bmi_',
    'polio',
    'adult_mortality',
    'alcohol',
    'hepatitis_b',
    'diphtheria_',
    'gdp',
    'population',
    '_thinness__1-19_years',
    '_thinness_5-9_years'
]

# Loop through each column in numerical_data_mean and fill missing values with the mean
for column in numerical_data_mean:
    df1[column] = df1[column].fillna(df1[column].mean())

# Loop through each column in numerical_data_median and fill missing values with the median
for column in numerical_data_median:
    df1[column] = df1[column].fillna(df1[column].median())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only the numerical columns for the correlation heatmap
numerical_df = df1.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap with Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True, cbar_kws={'shrink': .8})

# Display the heatmap
plt.title("Correlation Heatmap of Numerical Columns")
plt.show()
```

```python
# Group data by year and calculate the mean life expectancy for each year
yearly_life_expectancy = df1.groupby('year')['life_expectancy_'].mean()

# Plotting the trend of life expectancy over the years
plt.figure(figsize=(10, 6))
plt.plot(yearly_life_expectancy.index, yearly_life_expectancy.values, marker='o', color='purple')
plt.title('Life Expectancy Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Average Life Expectancy')
plt.grid(True)
plt.show()
```

```python
# Select the top 3 African countries with the highest life expectancy
top_lifea = african_countries.sort_values(by='life_expectancy_', ascending=True).head(3)

# Melt the data for the selected countries based on life expectancy factors
melted_data = top_lifea.melt(
    id_vars=['country'],
    value_vars=['adult_mortality', 'infant_deaths', 'schooling'],
    var_name='Factor',
    value_name='Score'
)

# Create the interactive bar plot with a pleasant color palette
fig = px.bar(
    melted_data,
    x='country',
    y='Score',
    color='Factor',
    barmode='group',
    title='Comparison of Life Expectancy Factors in Top African Countries',
    labels={'Score': 'Score', 'Country': 'Country'},
    color_discrete_sequence=px.colors.qualitative.Pastel  # Added a soft color palette
)

# Update layout for readability
fig.update_layout(
    xaxis_title='Country',
    yaxis_title='Score',
    legend_title='Life Expectancy Factors'  # Updated the legend title
)

# Show the interactive plot
fig.show()

```

**Sierra Leone, Angola, Central African Republic:**
Higher Adult Mortality Rates: Due to limited healthcare access, high rates of infectious diseases, malnutrition, and political instability.
**Factors:** HIV/AIDS, malaria, tuberculosis, and other preventable diseases significantly impact adult mortality.
In European Countries:

**Slovenia, Denmark, Cyprus:**
Lower Adult Mortality Rates: Advanced healthcare systems, high standards of living, and effective disease prevention contribute to lower adult mortality.

**Factors:** Better healthcare infrastructure and control over chronic diseases like heart disease and cancer.

#**2. Infant Deaths In African Countries:**

**Sierra Leone, Angola, Central African Republic:**
High Infant Mortality Rates: Limited access to healthcare, poor nutrition, and insufficient sanitation lead to higher rates of infant deaths.
Factors: Malnutrition, infectious diseases, lack of neonatal care, and poor maternal health are key contributors.
In European Countries:

**Slovenia, Denmark, Cyprus:**
Low Infant Mortality Rates: Access to quality prenatal and neonatal care, better nutrition, and sanitation reduce infant deaths.
Factors: Well-established healthcare systems, strong maternal health services, and high levels of hygiene.

# **3. Schooling In African Countries:**

**Sierra Leone, Angola, Central African Republic:**
Lower Schooling Rates: Due to economic challenges, limited access to education, and political instability, educational outcomes are lower.
Challenges: High dropout rates, insufficient educational resources, and lack of infrastructure in rural areas.
In European Countries:

**Slovenia, Denmark, Cyprus:**
Higher Schooling Rates: Strong educational systems, government investment in education, and widespread access to quality schooling lead to high literacy and graduation rates.
Factors: Compulsory education, well-funded schools, and a focus on lifelong learning opportunities.
